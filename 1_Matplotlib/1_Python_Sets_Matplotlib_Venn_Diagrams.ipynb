```python
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
from matplotlib_venn import venn3, venn3_circles
from matplotlib import pyplot as plt
%matplotlib inline
```

### Venn Diagram with 2 Groups

```python
plt.figure(figsize = (10, 8))
venn2(subsets = (a, b, d), set_labels = ('Group A', 'Group B'), alpha=0.8)
```

#### Customizing the Venn Diagram

- The default colors of the circles are red and green, but you can customize the colors.
- Add the set_colors parameter to set the colors of your groups. 
- In addition, you can use the alpha parameter to control the transparency.


```python
plt.figure(figsize = (10, 8))
venn2(subsets = (a, b, d), set_labels = ('Group A', 'Group B'), set_colors=('purple', 'skyblue'), alpha = 0.7);
```

- Venn diagram  automatically sizes the circle by the magnitude of the number of items within the circle. 
- However, if you do not wish for the circles to be weighted, you can use the unweighted Venn diagram instead.

```python
plt.figure(figsize = (10, 8))
venn2_unweighted(subsets = (a, b, d), set_labels = ('Group A', 'Group B'), set_colors=('r', 'g'), alpha = 0.5);
```

- You can further customize your Venn diagram by overlaying an outline of the circle over the original.
- Note that this works just with the weighted Venn diagrams.


```python
plt.figure(figsize = (10, 8))
venn2(subsets = (a, b, d),
      set_labels=('GroupA', 'GroupB'),
    set_colors=('r', 'g'), 
      alpha = 0.5)
venn2_circles(subsets = (a, b, d))
```

- If you do not prefer the solid line, you can change to different line style with the linestyle parameter.
- You can set the line width as well.

```python
plt.figure(figsize = (10, 8))
venn2(subsets = (a, b, d),
      set_labels=('GroupA', 'GroupB'),
    set_colors=('r', 'g'), 
      alpha = 0.5)
venn2_circles(subsets = (a, b, d), linestyle='dashed', linewidth=2, color= 'orange')
plt.title('I made a Venn Diagram!')
```

## Venn Diagram with 3 Groups

```python
P = {1, 2, 3, 4, 5, 6}
Q = {3, 5, 6, 8, 11, 13, 15, 16}
R = {6, 8, 12, 13, 14, 16, 17, 18, 19, 20}
```

```python
z = R.intersection(P.intersection(Q))
z
```

```python
z = len(z)
```

```python
X = P.union(Q)
Y = R.union(X)
Y
```

```python
y = len(Y)
```

```python
PP = len(P)
QQ = len(Q)
RR = len(R)
```

```python
pq = P.intersection(Q)
qr = Q.intersection(R)
rp = R.intersection(P)
```

```python
p = len(pq)
p
```

```python
q = len(qr)
q
```

```python
r = len(rp)
r
```

```python
plt.figure(figsize = (10, 8))
venn3(subsets = (PP,  QQ, p, RR, q, r, z), set_labels = ('Group P', 'Group Q', 'Group R'), alpha = 0.5);
```

- You can customize the colors of each area of the diagram with the get_patch_by_id method. 
- Notice that the middle area where all three circles overlap is now orange.

```python
plt.figure(figsize = (10, 8))
v = venn3(subsets = (PP,  QQ, p, RR, q, r, z), set_labels = ('Group P', 'Group Q', 'Group R'), alpha = 0.5);
v.get_patch_by_id('111').set_color('orange')   # id numbers 100, 110, 111
```

- How to change the line widths for each individual circle. If I set the Venn diagram circle to the variable c, 
- I can call each individual circle c[0], c[1], or c[2] and set the line width.

```python
c = venn3_circles(subsets = (20, 10, 12, 10, 9, 4, 3), linestyle='dashed', linewidth=1, color='grey')
c[0].set_lw(5.0)
c[1].set_lw(8.0)
c[2].set_lw(2.0)
```
